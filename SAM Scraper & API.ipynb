## Working in DataFrames

### JSON -> DataFrame function

```python
def json_to_dataframe(json_in):
    df = pd.DataFrame(json_in['opportunitiesData'])
    return df
```

```python
df = json_to_dataframe(response.json())
df
```

### A function that gets the contract description

```python
def get_desc(in_url):
    params = {'api_key':api_key.sam_api_key}
    response = requests.get(in_url, params = params)
    return response.json()['description']
```

```python
full_desc = []
for i in df.description:
    # print(get_desc(i).json()['description'])
    full_desc.append(get_desc(i))
print(full_desc)
```

```python
def get_all_desc(df_in):
    full_desc = []
    for i in df_in.description:
        full_desc.append(get_desc(i))
    return full_desc
```

### Add the new full descriptions to the DataFrame

```python
df['full_desc'] = get_all_desc(df)
df
```
